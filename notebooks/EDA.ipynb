```python
import pandas as pd
import ploty.graph_objects as go
```

```python
data = pd.read_csv("data/vehicles.csv")
data.head()
```

```python
# histograma con go
fig = go.Figure(data=[go.Histogram(x=data['year'], nbinsx=20)])
fig.update_layout(title='Distribución de años de fabricación de vehículos',
                  xaxis_title='Año de fabricación',
                  yaxis_title='Cantidad de vehículos')
fig.show()
```
